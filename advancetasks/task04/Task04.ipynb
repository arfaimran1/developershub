```python
import gradio as gr
import numpy as np
from sentence_transformers import SentenceTransformer
import faiss
```

```python
documents = [
    "Artificial Intelligence is the simulation of human intelligence in machines.",
    "Machine Learning is a subset of AI that learns patterns from data.",
    "Deep Learning uses neural networks with many layers.",
    "Python is a popular programming language for AI and data science.",
    "Data Science combines statistics and machine learning to analyze data."
]
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')

doc_embeddings = model.encode(documents)
```

```python
dimension = doc_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(doc_embeddings))
```

```python
def chatbot(query):
    query_embedding = model.encode([query])

    distances, indices = index.search(np.array(query_embedding), k=1)

    best_match = documents[indices[0][0]]

    return f"Answer: {best_match}"
```

```python
interface = gr.Interface(
    fn=chatbot,
    inputs=gr.Textbox(label="Ask your question"),
    outputs="text",
    title="🤖 Context-Aware Chatbot (Task 4)",
    description="RAG-style chatbot using embeddings + similarity search"
)

interface.launch()
```
